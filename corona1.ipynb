```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


#Reading data
data1= pd.read_csv("/content/corona_tested_006.csv")
print(data1.shape)
data1.head()

#Total number of values
print("Number of patients", str(len(data1.index)))

#Data Analysis
#Patients who have corona and patients who do not have corona
sns.countplot(x="Corona", data= data1)


```

```python
#Of who did not have corona , how many had cough and how many did not have corona
sns.countplot(x="Cough", hue="Cough_symptoms", data= data1)
```

```python
#Of who did not have Corona , how many had fever and how many did not have fever
sns.countplot(x="Fever", hue="Fever", data= data1)
```

```python
#Of who did not have Corona , how many had sore throat and how many did not have sore throat
sns.countplot(x="Sore Throat", hue="Sore_throat", data= data1)
```

```python
#Of who did not have Corona , how many had shortness of breath and how many did not have it
sns.countplot(x="Corona", hue="Shortness_of_breath", data= data1)
```

```python
#Of who did not have Corona , how many had Head ache and how many did not have Head ache
sns.countplot(x="Corona", hue="Headache", data= data1)
```

```python
#Of who did not have Corona , how many are above 60 and not
sns.countplot(x="Corona", hue="Age_60_above", data= data1)
```
